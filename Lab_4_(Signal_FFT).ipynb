```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/Advanced Lab 1/Lab 4/Lab 4 Data.csv'
data = np.genfromtxt(path, delimiter=',', names=True)
```

```python
time = data['Time']
time = time/1000
voltage = data['Voltage']
```

```python
fig, ax = plt.subplots()

ax.plot(time, voltage, label="thermistor")

ax.set_xlabel('Time [minutes]')
ax.set_ylabel('Temprature Measured by Thermistor [K]')
ax.set_title('Signal as a function of time')

ax.grid()
plt.show()
```

```python
from scipy.fft import fft, fftfreq
from scipy.fft import rfft, rfftfreq
```

```python
N = len(voltage)
sample_rate = 1000

yf = fft(voltage)
xf = fftfreq(N, 1 / sample_rate)
fig, ax = plt.subplots()
ax.plot(xf, np.abs(yf),color='blue')
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Signal')
ax.set_title('Signal FFT Spectrum')
ax.set_xlim(0,10)
ax.set_ylim(0,1e6)
ax.grid()
plt.show()
```
