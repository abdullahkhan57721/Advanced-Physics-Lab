```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sci
```

```python
path1 = "/content/drive/MyDrive/Colab Notebooks/Advanced Lab 1/Project/Closed Sweep.csv"
data1 = np.genfromtxt(path1, delimiter=',', names=True)
```

```python
time = data1["Time"]/100000*900
signal = data1["Signal"]
```

```python
peaks,_ = sci.find_peaks(signal, height=669, distance=15000)
peaks_avg = np.mean(time[peaks])

fig, ax = plt.subplots(figsize=(6,3), dpi=150)

ax.plot(time, signal, color='orange', label="Waveform")

ax.plot(time[peaks], signal[peaks], "x")
for x in time[peaks]:
  plt.axvline(x, color = 'b', linewidth=0.5)

ax.set_xlabel('Frequency Shifted by a Constant [Hz]')
ax.set_ylabel('Signal Amplitude [AU]')
ax.set_title('Closed Tube Resonance Amplification of a 100-1000 Hz Frequency Sweep')
ax.grid()
plt.xticks(time[peaks])
plt.yticks(np.arange(min(signal), max(signal)+1, 2.5))
plt.legend()
plt.show()
```
