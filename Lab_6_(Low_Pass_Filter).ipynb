```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd;
```

```python
path = "/content/drive/MyDrive/Colab Notebooks/Advanced Lab 1/Lab 6/Signal Amplitude vs Frequency.csv"
data = np.genfromtxt(path, delimiter=',', names=True)
```

```python
frequency = data["Frequency"]
signal = data["Signal"]
signal = signal
signal = np.sqrt(np.square(signal))
signal = 20*np.log(signal/950)
theory = np.sqrt(1/(1+(2*np.pi*22*0.0001*frequency)**2))
theory = 20*np.log(theory)


fig, ax = plt.subplots(figsize=(6,3), dpi=150)


ax.vlines(x = 80, ymin = -55, ymax = 0, color = 'purple', label="Cutoff frequency")
ax.plot(frequency, theory, color='black', label="Theoretical prediction")
# setting scaley = True will make the line fit
# withn the frame, i.e It will appear as a finite line

ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('RMS Voltage [dB]')
ax.set_title('Filtered Signal Strength as a Function of Frequency')
plt.xscale("log")

plt.legend()
plt.show()
```
