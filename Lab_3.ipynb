```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/Advanced Lab 1/Lab 3/Data_Lab_3.csv'
data = np.genfromtxt(path, delimiter=',', names=True)
```

```python
time = data['time']
time = time/60000
potValue = data['potValue']
potValue = potValue/1024*5
sensorValue = data['sensorValue']
sensorValue = sensorValue/1024*5
```

```python
resistance = 3.3*((5/sensorValue)-1)
a = 3.354016e-3
b = 2.569850e-4
c = 2.620131e-6
d = 6.383091e-8
ref = 3.3
temprature = 1/(a + b*np.log(resistance/ref) + c*np.log((resistance/ref)**2) + d*np.log((resistance/ref)**3))
```

```python
fig, ax = plt.subplots()
ax2 = ax.twinx()

ax.plot(time, temprature, 'g', label="thermistor")
ax2.plot(time, potValue, 'b', label="potentiometer")

ax.set_xlabel('Time [minutes]')
ax.set_ylabel('Temprature Measured by Thermistor [K]')
ax2.set_ylabel('Potentiometer Voltage [V]')
ax.set_title('Temprature as a function of time')

#legend positioning code via https://stackoverflow.com/a/47370214
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax.transAxes)

ax.grid()
plt.show()
```
